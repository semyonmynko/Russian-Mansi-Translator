```python
import requests
from lxml import html
import re
import json
import pandas as pd
```

```python
# Ссылка на страницу словаря
def get_link(page):
    if page < 10:
        page = f'0{page}'
    return f'http://finugorbib.com/bible/mansi/41_Mar{page}_ru.html'

# Количество страниц (выявлено в ручную)
page_limit = 17

# Словари для слов
list_lang_mans = []
list_lang_rus = []
```

```python
# Парсинг каждой страницы
print("Парсинг успешно начат!")
for page in range(1, page_limit):
    print(f">>> Парсинг {page} страницы:")

    response = requests.get(url=get_link(page))
    tree = html.fromstring(re.sub(r'<\s*a[^>]*>(.*?)<\s*/\s*a>', '', response.text))

    # Посчитать количество элементов на странице
    count_elements = len(tree.xpath('.//div[@id="middle"]//table//tr'))

    # Для каждого перевода
    for i in range(3, count_elements + 1):
        sentence_mans = tree.xpath(f'.//div[@id="middle"]//table//tr[{i}]//td[1]/text()')[0]
        sentence_rus = tree.xpath(f'.//div[@id="middle"]//table//tr[{i}]//td[2]/text()')[0]

        # Сохранить элементы в соответствующие словари
        list_lang_mans.append(sentence_mans)
        list_lang_rus.append(sentence_rus)

        print(f"    Добавлен перевод: {sentence_mans} — {sentence_rus})")

print("Парсинг успешно завершён!")
```

```python
print("Парсинг успешно завершён!")
# Создание DataFrame
df = pd.DataFrame(
    {'target': list_lang_mans,
     'source': list_lang_rus
     }
)
```

```python
# Сохранение DataFrame to File
df.to_csv("dataset_bible.csv", sep="|")
```

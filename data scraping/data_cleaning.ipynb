```python
import pandas as pd
import re
```

## Cleaning parallel corpus
Removing duplicates, detection of words written l i k e t h i s and fixing it, removing nan rows, etc.

```python
df = pd.read_csv('data/overall_80K.csv')
df
```

```python
df = df.drop(columns=df.columns[0])
```

```python
df = df.drop_duplicates()
df
```

```python
def clean_text(text):
    # Remove URLs
    text = re.sub(r"http\S+", "", text)
    # Remove HTML tags
    text = re.sub(r"<.*?>", "", text)
    # Replace any kind of space (including non-breaking spaces) with a standard space
    text = re.sub(r"\s+", " ", text)
    # Strip leading/trailing spaces
    return text.strip()
```

```python
df['target'] = df['target'].apply(clean_text)
df
```

```python
# \b\w\b  --> Matches any single letter (\w) that is a whole word (\b on both sides).
pattern = r'\S\s\S\s\S\s\S\s\S\s'
# Use df['target'].str.contains() to filter rows matching the pattern
df = df[~df['target'].str.contains(pattern, regex=True)]
df
```

```python
df.dropna().replace('', pd.NA).dropna() # No empty values
```

```python
df.to_csv('data/overall_80K_cleaned.csv')
```

## Collecting and cleaning of monolingual mansi coprus

```python
horvat = pd.read_csv('data\mansi_from_horvath_mono.csv')
horvat = horvat.drop(columns=df.columns[0])
```

```python
mns = pd.read_csv('data\luima_seripos_mansi.csv')
mns = mns.drop(columns=df.columns[0])
mns
```

```python
mns = pd.concat([mns, horvat], ignore_index=True)
```

```python

mns = mns.dropna().replace('', pd.NA).dropna() 
mns = mns.drop_duplicates()
mns = mns[mns.columns[0]]
mns
```

```python
mns = mns.apply(clean_text)
```

```python
mns = mns[mns.str.len() >= 20]
mns.tail(20)
```

```python
mns.to_csv('data/mansi.csv')
```
